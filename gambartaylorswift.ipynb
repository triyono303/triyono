```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from google.colab import files
uploaded = files.upload()
```

```python
# Resizing 0.5

img=cv2.imread('TaylorSwift.JPG')
img2=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img3=cv2.resize(img2,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)
plt.imshow(img3)
```

```python
# Resizing 2
height, width=img.shape[:2]
res=cv2.resize(img2,(2*width,2*height),interpolation=cv2.INTER_CUBIC)
plt.imshow(res)
```

```python
# Translation
rows,cols=img2.shape[:2]
M=np.float32([[1,0,100],[0,1,50]])
dst=cv2.warpAffine(img2,M,(cols,rows))
plt.imshow(dst)
```

```python
# Rotation 90
M=cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst=cv2.warpAffine(img2,M,(cols,rows))
plt.imshow(dst)
```

```python
# Affine transformation
rows,cols,ch=img2.shape
pts1=np.float32([[50,50],[200,50],[50,200]])
pts2=np.float32([[10,100],[200,50],[100,250]])
M=cv2.getAffineTransform(pts1,pts2)
dst=cv2.warpAffine(img2,M,(cols,rows))
plt.imshow(dst)
```

```python
# Perspective transform
pts1=np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2=np.float32([[0,0],[300,0],[0,300],[300,300]])
M=cv2.getPerspectiveTransform(pts1,pts2)
dst=cv2.warpPerspective(img2,M,(300,300))
plt.imshow(dst)
```
